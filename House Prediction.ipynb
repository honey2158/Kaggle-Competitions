```python
cat_vars = train.select_dtypes(include=["O"]).columns
cat_vars
#len(cat_vars)
```

```python
from sklearn.impute import KNNImputer  

def impute_numerical(df, columns):    
    imputer = KNNImputer(n_neighbors=5)     
    df[columns] = imputer.fit_transform(df[columns])    
    return df

train = impute_numerical(train, num_vars)
```

```python
def impute_categorical(df, columns):
    
    df_onehot_cat = pd.get_dummies(df[columns], drop_first = True,)    
    df = df.drop(columns = columns)      
    df = pd.concat([df, df_onehot_cat], axis=1)      
    return df

train = impute_categorical(train, cat_vars)
```

```python
from sklearn.preprocessing import StandardScaler

def normalize(df):
    scaler = StandardScaler()  # Initiate the scaler
    columns = df.columns.tolist()
    #columns.remove(exclude)
    df[columns] = pd.DataFrame(data=scaler.fit_transform(df[columns]), columns=columns)
    return df
train = normalize(train)
```

```python
X = train[features_train]
y = Y
```

```python
len(X.columns)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.25)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier_nb = GaussianNB()
classifier_nb.fit(X_train, y_train)
y_pred = classifier_nb.predict(X_test)

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier_dtc = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier_dtc.fit(X_train, y_train)
y_pred = classifier_dtc.predict(X_test)

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier_rfc = RandomForestClassifier(n_estimators = 100, criterion = 'entropy')
classifier_rfc.fit(X_train, y_train)
y_pred = classifier_rfc.predict(X_test)

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier_knn = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier_knn.fit(X_train, y_train)
y_pred = classifier_knn.predict(X_test)

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
GBRegressor = GradientBoostingRegressor(random_state=42)  
GBRegressor.fit(X_train, y_train)
GBRegressor.score(X_test, y_test)
```
